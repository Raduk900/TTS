```python
import soundfile as sf
import noisereduce as nr
import librosa
import IPython.display as ipd
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
# Load audio file
audio_data, sample_rate = librosa.load('dataset/LJSpeech-1.1/wavs/28a595c8332139b97a12c8bf0bcd6583.wav', sr=None)

noise_clip = audio_data[0:10000]

reduced_noise_audio = nr.reduce_noise(y=audio_data, sr=sample_rate)

sf.write('your_audio_file_reduced_noise.wav', reduced_noise_audio, sample_rate)
```

```python
y, sr = librosa.load('your_audio_file_reduced_noise.wav', sr=None)
duration_in_seconds = len(y) / sr
time_in_milliseconds = pd.Series(range(len(y))) * (duration_in_seconds / len(y)) * 1000
audio_series = pd.Series(y, index=time_in_milliseconds)
```

```python
audio_series.plot(figsize=(10, 5), lw=1, title='Audio file')
plt.xlabel('Time (milliseconds)')
plt.ylabel('Amplitude')
plt.show()
```
